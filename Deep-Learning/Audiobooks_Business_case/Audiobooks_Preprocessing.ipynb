# Audiobooks business case

### Extract the data from the csv

```python
import numpy as np
from sklearn import preprocessing
```

```python
data = np.loadtxt('Audiobooks_data.csv', delimiter=',')

variable_data = data[:,1:-1]
target = data[:,-1]
```

### Balance the dataset

```python
no_of_ones = int(np.sum(target))
no_of_zeros = 0

indices_to_remove = []

for i in range(target.shape[0]):
    if target[i] == 0:
        no_of_zeros += 1
        if no_of_zeros > no_of_ones:
            indices_to_remove.append(i)

unscaled_input_equal_priors = np.delete(variable_data,indices_to_remove,axis=0)
targets_equal_priors = np.delete(target,indices_to_remove,axis=0)
```

### Standardize the inputs

```python
scaled_inputs = preprocessing.scale(unscaled_input_equal_priors)
```

### Shuffle the data

```python
shuffled_indices = np.arange(scaled_inputs.shape[0])
np.random.shuffle(shuffled_indices)

inputs = scaled_inputs[shuffled_indices]
targets = targets_equal_priors[shuffled_indices]
```

### Split the dataset into train, validation, and test

```python
data_size = inputs.shape[0]
```

```python

train_size = int(0.8*data_size)
test_size = int(0.1*data_size)
validation_size = int(0.1*data_size)

train_inputs = inputs[:train_size,:]
train_targets = targets[:train_size]

validation_input = inputs[train_size:train_size+validation_size,:]
validation_targets = targets[train_size:train_size+validation_size]

test_input = inputs[train_size+validation_size:,:]
test_targets = targets[train_size+validation_size:]
```

```python
# Checking if data is balanced

print(np.sum(train_targets), train_size, np.sum(train_targets) / train_size)
print(np.sum(validation_targets), validation_size, np.sum(validation_targets) / validation_size)
print(np.sum(test_targets), test_size, np.sum(test_targets) / test_size)
```

### Save the three datasets in *.npz

```python
np.savez('Audiobooks_data_train',inputs=train_inputs,targets=train_targets)
np.savez('Audiobooks_data_validation',inputs=validation_input,targets=validation_targets)
np.savez('Audiobooks_data_test',inputs=test_input,targets=test_targets)
```
