## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Generating random data

```python
obs = 100000

xs = np.random.uniform(-10,10,(obs,1))
zs = np.random.uniform(-10,10,(obs,1))
```

```python
inp = np.column_stack((xs,zs))
inp.shape
```

## Target

```python
noise = np.random.uniform(-1,1,(obs,1))

targets = 2*xs + 3*zs +5 + noise

targets.shape
```

## Initialize Variables

```python
init_range = 0.1

weights = np.random.uniform(-init_range,init_range,(2,1))
bias = np.random.uniform(-init_range,init_range,1)
```

## Learning Rate

```python
learning_rate = 1
```

## Train The model

```python
for i in range(100):
    output = np.dot(inp,weights) + bias

    delta = output - targets

    loss = np.sum(delta**2)/2/obs
    print(loss)

    delta_scaled = delta / obs
    weights = weights - learning_rate*np.dot(inp.T,delta_scaled)
    bias = bias - learning_rate*np.sum(delta_scaled)
```
