# Imports

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing  import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from  sklearn import metrics
```

# Data

```python
df = pd.read_csv('Absenteeism_preprocessed.csv')
df.head()
```

## Set target

```python
targets = np.where(df['Absenteeism Time in Hours']<=np.median(df['Absenteeism Time in Hours']),0,1)
targets
```

```python
df['Excessive Absenteeism'] = targets
df.head()
```

### Checkpoint

```python
df_with_targets = df.drop(['Absenteeism Time in Hours'],axis=1)
```

```python
df_with_targets.head()
```

## Input

```python
unscaled_inputs = df_with_targets.iloc[:,:-1]
```

# Standardization

```python
columns_to_scale = ['Month',                     'Day',
  'Transportation Expense',        'Distance to Work',
                     'Age', 'Daily Work Load Average',
         'Body Mass Index',               'Education',
                'Children',                    'Pets']
```

```python

Absenteeism_scaler = ColumnTransformer(transformers=[('scaler',StandardScaler(),columns_to_scale)],remainder='passthrough')

scaled_inputs = Absenteeism_scaler.fit_transform(unscaled_inputs)

scaled_inputs
```

# Train Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs,targets,test_size=0.2,random_state=20)
```

# Model

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)*100
```

# Summary

```python
model.intercept_
```

```python
model.coef_
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary = pd.DataFrame(columns=['feature name'],data = feature_name)
```

```python
summary['Coeficients'] = np.transpose(model.coef_)
```

```python
summary.index = summary.index+1
```

```python
summary.loc[0] = ['Intercept',model.intercept_[0]]
summary = summary.sort_index()
```

```python
summary
```

```python
summary['Odds Ratio'] = np.exp(summary['Coeficients'])
```
