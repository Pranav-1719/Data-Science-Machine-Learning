# Imports

```python
import pandas as pd
```

```python
raw_data = pd.read_csv('Absenteeism_data.csv')

df = raw_data.copy()
df
```

```python
df.info()
```

```python
reason_dummies = pd.get_dummies(df['Reason for Absence'],drop_first=True,dtype=int)
reason_dummies['Sum']= reason_dummies.sum(axis=1)
reason_dummies['Sum'].unique()
```

## Splitting Reasons into 4 groups

```python
reason_type1 = reason_dummies.loc[:,:14].max(axis=1)
reason_type2 = reason_dummies.loc[:,15:17].max(axis=1)
reason_type3 = reason_dummies.loc[:,18:21].max(axis=1)
reason_type4 = reason_dummies.loc[:,22:].max(axis=1)
```

```python
df = df.drop(['ID','Reason for Absence'],axis=1)
df.head()
```

```python
df = pd.concat([df,reason_type1,reason_type2,reason_type3,reason_type4],axis=1)
```

```python
df.columns = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2','Reason_3' ,'Reason_4']
```

```python
col_order = [ 'Reason_1', 'Reason_2','Reason_3' ,'Reason_4',
       'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df = df[col_order]

df.head()
```

## Checkpoint

```python
df_checkpoint = df.copy()
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d/%m/%Y')

type(df['Date'][0])
```

```python
df.head()
```

```python
list_month=[]
```

```python
for i in range(df.shape[0]):
    list_month.append(df['Date'][i].month)
```

```python
df['Month'] = list_month
```

```python
def get_day(date_value):
    return date_value.weekday()

df['Day'] = df['Date'].apply(get_day)
```

```python
df.head(20)
```

```python
df = df.drop(['Date'],axis=1)
```
